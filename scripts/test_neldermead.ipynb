```python
import numpy as np
import time
import matplotlib.pyplot as plt

from bezierv.classes.bezierv import Bezierv
from bezierv.algorithms.nelder_mead import NelderMeadSolver
```

```python
#data
x_sample = np.sort(np.random.normal(0, 1, 200))
```

```python
n = 3
controls_x0 = np.linspace(x_sample[0] -1 , x_sample[-1] + 1, n + 1)
controls_z0 = np.linspace(0, 1, n + 1)
bezierv_subgrad = Bezierv(n)
nelder_mead = NelderMeadSolver(bezierv_subgrad, x_sample)
fitted = nelder_mead.solve(controls_x0=controls_x0, controls_z0=controls_z0)
```

```python
fitted.plot_cdf(x_sample)
```
