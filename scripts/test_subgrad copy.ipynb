```python
import numpy as np
import time
import matplotlib.pyplot as plt

from bezierv.classes.bezierv import Bezierv
from bezierv.algorithms.non_linear_solver import NonLinearSolver
from bezierv.algorithms.proj_grad import ProjGrad
from bezierv.algorithms.proj_subgrad import ProjSubgrad
```

```python
#data
x_sample = np.sort(np.random.normal(0, 1, 500))
```

```python
n = 5
controls_z0 = np.linspace(0, 1, n + 1)
bezierv_subgrad = Bezierv(n)
proj_subgrad = ProjSubgrad(bezierv_subgrad, x_sample, emp_cdf_data=None)
fitted = proj_subgrad.fit(controls_z0=controls_z0, controls_x0=None, maxiter=100, step=1e-2)
```

```python
fitted.plot_cdf(x_sample)
print(fitted.controls_x)
print(proj_subgrad.get_controls_x(x_sample))
```
